## Question 2 - 2 Points

Steam records the average number of players per game each month. For this assignment, we are analyzing the popularity of the game **Counter Strike: Global Offensive**, a competitive FPS game. Create a function that randomly picks one of the games in each of the 12 months, then simulate this selection 2,000 times. Save the results in a table called **simulations**, and display the first 10 rows of the table.

**Hint:** The simulations table should have 12 columns, each representing one of the different months. A list of the months has been provided.

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
# Place answer here.

import random


def get_month(table):
    games = table.column("Game_Name")
    return random.choice(games)

simulations = Table()
for month in months:
    rows = make_array()
    month_table = steam_2020.where("Month", month)
    for i in range(2000):
        assigned = get_month(month_table)
        rows = np.append(rows, assigned)
    simulations = simulations.with_column(month, rows)


simulations.show(10)
```

## Question 3a - 2 Points

Using the **simulations** table created in Question 2, display a histogram showing the distribution of months Counter Strike: Global Offensive was the top ranked game. The bins for the histogram should be the number of months (0-12) where Counter Strike: Global Offensive was the top ranked game that specific number of times.  Make the histogram as readable as possible.

```python


months_idx = []
for i in range(simulations.num_rows):
    games = np.array(simulations.row(i))
    count = np.count_nonzero(games == "Counter Strike: Global Offensive")
    months_idx.append(count)
cs = Table().with_column("Count",months_idx)

cs.hist("Count", bins=np.arange(-0.5,13,1))
plots.title("Number of Times that Counter Strike: Global Offensive was the Simulated Top Ranked Game")
plots.xlabel("Number of Months")
plots.ylim(-.01)
plots.show()

```

## Question 3d - 1 Point

Calculate a p-value that uses 12 as the **observed statistic**. Display the value as a percentage 
with 1 digit to the right of the decimal, e.g. "*The p-value is 0.3%*"

```python
observed_statistic = 12
percent = 100 * (np.count_nonzero(cs.column("Count") >= observed_statistic) / simulations.num_rows)
print(f"The p-value is {percent:.1f}%")
```

Based on the calculated p-value, should the **null hypothesis** be accepted or rejected?  Explain your answer.

**Answer -** With a p-value of 0.0% we would reject the null hypothesis. This is because under the assumptions of the null hypothesis, it is extremely unlikely to get a value as large as the observed statistic that we actually saw. This provides strong evidence that Counter Strike: Global Offensive has some advantage over other games. It appears as #1 ranked game of the month too many times to be considered chance. 

## Question 4 - 2 Points

Discover something insightful with the provided data set. Explain what you have done and why it is insightful.

```python

top_games = steam_games.where("monthly_rank", 1)
abv_months = top_games.with_column("Abv Month", [m[:3] for m in top_games.column("Month")])
top_2019 = abv_months.where("Year", 2019)
top_2020 = abv_months.where("Year", 2020)

plots.plot(top_2019.column("Abv Month"), top_2019.column("Avg_players"), label="2019")
plots.plot(top_2020.column("Abv Month"), top_2020.column("Avg_players"), label="2020")
plots.xlabel("Month")
plots.ylabel("Number of Average Monthly Players")
plots.legend()
plots.title("Average Players for #1 Game of the Month")
plots.show()
```

**Answer -** This plot shows the increase in average monthly players for that month's number one game. Although this is not a representation of the total number of people playing video games, we can observe a large increase in average players in 2020. There is a major spike beginning in March 2020 with player counts continuing to stay a consistent 100,000-200,000 higher than the previous year. This appears to be consistent with the COVID-19 lockdown timeline. One thing to consider is that this graph only shows the top game for each month and could easily be influenced by the release of a popular game.
